7-What is the difference between GET and POST HTTP methods?
 1. Purpose
* GET: Used to retrieve data from the server.

    * POST: Used to send or submit data to the server.

 2. Data Transmission
* GET: Data is sent via the URL (as query parameters).

  * POST: Data is sent in the body of the request.

3. Visibility
 * GET: Data is visible in the browser‚Äôs address bar.

 * POST: Data is not visible in the URL.

4. Security
 * GET: Less secure, not recommended for sensitive data.

 *  POST: More secure for sending passwords or form data.

5. Caching
 * GET: Can be cached by browsers.

 * POST: Not cached by default.

6. Use Cases
 * GET: Reading or fetching data (e.g., view user profile).

 * POST: Creating or submitting data (e.g., login form, registration).

 * GET: GET /search?query=flask

 * POST: POST /register with body:

2- How do you serve static files like images or CSS in Flask?

```python
# üìÅ Project Structure:
# your_project/
# ‚îú‚îÄ‚îÄ app.py
# ‚îú‚îÄ‚îÄ static/
# ‚îÇ   ‚îú‚îÄ‚îÄ style.css
# ‚îÇ   ‚îî‚îÄ‚îÄ image.jpg
# ‚îî‚îÄ‚îÄ templates/
#     ‚îî‚îÄ‚îÄ index.html

# üî∏ app.py
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    return render_template('index.html')

if __name__ == '__main__':
    app.run(debug=True)

# üî∏ static/style.css
# (Save this in a file named style.css inside the 'static' folder)
"""
body {
    background-color: #f0f0f0;
    font-family: Arial, sans-serif;
    text-align: center;
    padding: 50px;
}
h1 {
    color: #333;
}
img {
    width: 300px;
    margin-top: 20px;
}
"""

# üî∏ templates/index.html
# (Save this in a file named index.html inside the 'templates' folder)
"""
<!DOCTYPE html>
<html>
<head>
    <title>Flask Static Files</title>
    <link rel="stylesheet" href="{{ url_for('static', filename='style.css') }}">
</head>
<body>
    <h1>Welcome to Flask!</h1>
    <img src="{{ url_for('static', filename='image.jpg') }}" alt="Sample Image">
</body>
</html>
"""
```

3- How do you define different routes with different HTTP methods in Flask?

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# GET route
@app.route('/hello', methods=['GET'])
def hello():
    return 'Hello, GET request!'

# POST route
@app.route('/submit', methods=['POST'])
def submit():
    data = request.get_json()
    return jsonify({'message': 'Data received', 'data': data}), 201

# PUT route
@app.route('/update/<int:id>', methods=['PUT'])
def update(id):
    data = request.get_json()
    return jsonify({'message': f'User {id} updated', 'data': data})

# DELETE route
@app.route('/delete/<int:id>', methods=['DELETE'])
def delete(id):
    return jsonify({'message': f'User {id} deleted'}), 200

if __name__ == '__main__':
    app.run(debug=True)
```

5- How can you generate URLs for routes in Flask using url_for?

```python
# üìÅ Project Structure:
# your_project/
# ‚îú‚îÄ‚îÄ app.py
# ‚îî‚îÄ‚îÄ templates/
#     ‚îî‚îÄ‚îÄ index.html

# üî∏ app.py
from flask import Flask, render_template, redirect, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/about')
def about():
    return "This is the About Page"

@app.route('/redirect-to-about')
def go_to_about():
    return redirect(url_for('about'))  # Generates /about URL

if __name__ == '__main__':
    app.run(debug=True)

# üî∏ templates/index.html
# (Save this inside templates/index.html)
"""
<!DOCTYPE html>
<html>
<head>
    <title>url_for Example</title>
</head>
<body>
    <h1>Welcome!</h1>
    <a href="{{ url_for('home') }}">Home</a> |
    <a href="{{ url_for('about') }}">About</a> |
    <a href="{{ url_for('go_to_about') }}">Go to About (Redirect)</a>
</body>
</html>
"""
```

6-  How do you handle forms in Flask?

```python
# üìÅ Project Structure:
# your_project/
# ‚îú‚îÄ‚îÄ app.py
# ‚îî‚îÄ‚îÄ templates/
#     ‚îî‚îÄ‚îÄ form.html

# üî∏ app.py
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def form():
    if request.method == 'POST':
        name = request.form['name']
        return f"<h2>Hello, {name}!</h2>"
    return render_template('form.html')

if __name__ == '__main__':
    app.run(debug=True)

# üî∏ templates/form.html
# (Save this as templates/form.html)
"""
<!DOCTYPE html>
<html>
<head>
    <title>Flask Form</title>
</head>
<body>
    <h1>Enter Your Name</h1>
    <form method="POST">
        <input type="text" name="name" required>
        <button type="submit">Submit</button>
    </form>
</body>
</html>
"""
```

7- How can you validate form data in Flask?

```python
from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def form():
    error = None
    if request.method == 'POST':
        name = request.form.get('name')
        if not name:
            error = "Name is required!"
        elif len(name) < 3:
            error = "Name must be at least 3 characters."
        else:
            return f"Hello, {name}!"
    return render_template('form.html', error=error)

if __name__ == '__main__':
    app.run(debug=True)
```

12- How do you define a custom Jinja filter in Flask?

```python

from flask import Flask, render_template

app = Flask(__name__)

# üîπ Define a custom Jinja filter
def reverse_string(s):
    return s[::-1]

# üîπ Register the filter
app.jinja_env.filters['reverse'] = reverse_string

@app.route('/')
def home():
    name = "Naveen"
    return render_template('index.html', name=name)

if __name__ == '__main__':
    app.run(debug=True)
```

13- How can you redirect with query parameters in Flask?

```python
from flask import Flask, redirect, url_for, request

app = Flask(__name__)

@app.route('/')
def index():
    # Redirect to /welcome with query parameter ?name=Naveen
    return redirect(url_for('welcome', name='Naveen'))

@app.route('/welcome')
def welcome():
    # Access the query parameter
    name = request.args.get('name')
    return f"Welcome, {name}!"

if __name__ == '__main__':
    app.run(debug=True)
```
